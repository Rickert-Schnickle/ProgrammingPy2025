### Calculations, Variables, Operators, and Comments

First things first, let's talk about some basics of Python. As you have seen in the lecture, its most straightforward use is as a calculator. Type any calculation into the code block below, hit Ctrl+Enter, and Jupyter will produce a result (try it!)

```python
## Code block 2: Python as a calculator

1 + 1
```

Note: one of the nice things with Jupyter notebooks is that just typing the name of a variable, or a calculation is enough to print the result (= evaluation) below the code block (also called "printing to the console"). When you are working with a different format (e.g., Python files), this will not work. In that case, you need to put what you want to print inside a print-statement, i.e. `print()`:


```python
## Code block 3: Python as a calculator with print statement

print(1 + 1)
```

As you see, code block 3 outputs the exact same thing as code block 2. This print statement is not necessary when you only want to print one evaluation in your code block, but oftentimes we want to print more than one thing, so it is good to familiarize yourself with the idea that you write a print statement around what you want to print.

You can also store things in variables, using the `=` operator. The name of the variable goes to the left of the equals-sign, its contents go to the right. Then if you want to see what is stored in the variable, type its name in the code block. For example:

```python
## Code block 3: Variables

my_var = "a"
print(my_var)
```

This has put the character `"a"` inside the variable `my_var`. You can store anything as a variable: integers, floating-point numbers, strings, `True` and `False`, as well as more complex data structures, like lists and functions, that we will get to later. Feel free to try to store other values in the `my_var` variable in the code block above. You can change an object's type at will; re-using it just overwrites it:

```python
## Code block 4: Overwriting a variable

my_var = 1
print(my_var)
my_var = True
print(my_var)
```

Also, if you like, you can try to change the values in `my_var` in the code block above and remove the `print()` statements, to see what happens if you don't use those.

If you want Python to store a string like `"a"` or maybe `"apple"`, you have to surround it in quotation marks (both single and double quotation marks work; some people have very strong opinions on what is best, I don't). Otherwise Python thinks you're referring to the name of an object. So this will give an error:

```python
## Code block 5: Strings require quotation marks

my_var = apple
```

If you see errors like this in the code block examples, try to make a mental note of it and what it means. NameError in this case means that we are trying to work with a variable (`apple`) that has not been created. In the exercises you may come across similar errors (which, again, is totally normal!) and it is very useful to have an idea of what they mean and how to solve them.

You can give basically any name to variables, but it is always good to choose a descriptive name that helps you and others understand what is stored (think of names like `age`, `car_brand`, `musician_name`, etc.). There are a few rules though:

- A variable name must start with a letter or the underscore character
- A variable name cannot start with a number
- A variable name can only contain alpha-numeric characters and underscores (`A-z`, `0-9`, and `_` )
- Variable names are case-sensitive (`age`, `Age` and `AGE` are three different variables)

## Comparison operators, Boolean values, and flow control statements

In many cases, we do not want to just perform basic calculations and modifications with Python. We want to use this calculated/modified information and process it in some way. For instance, by comparing values to each other. That's where comparison operators in combination with Boolean values come into play. As said in the lecture, there are two Boolean values: `True` or `False`, and comparisons that we are making will boil down to either one of them. Let's check it out.


```python
## Code block 6: Boolean comparisons

firstcomparison = 4 > 3
print(firstcomparison)

secondcomparison = 4 < 3
print(secondcomparison)
```

In the code block above, we first ask _"Is the value `4` more than the value `3`?"_, for which the answer is _"yes"_, so it takes on the Boolean value `True`. We than ask _"Is the value `4` less than the value `3`?"_, which is not the case, of course: `False`.

We can extend the amount of processing we do even further, by using _Flow control statements_. Then, we don't just want our yes/no (`True`/`False`) answer, but we also want to do something with that `True`/`False` answer. This is done with `if`, `elif`, and `else`. In this worksheet, we will focus on the `if` and `else` statements. Let's compare the following code block to code block 6.

```python
## Code block 7: Flow control statements

if 4 > 3:
    print('4 is indeed more than 3')
else:
    print('No, 4 is less than 3')
```

As you may be able to tell, code block 7 is basically an extension of the `firstcomparison` variable of code block 6. But rather than printing (`True`/`False`), we print a more detailed string. What this code block basically does is it starts again with the question _"Is the value `4` more than the value `3`?"_, but then extends it with an _"If so, print `4 is indeed more than 3` to the console (the `if`-statement), but if that is not the case, print `No, 4 is less than 3` to the console (the `else` statement). Pay special attention to the spacing that is applied here. When you are working with flow control statements, you want everything that belongs to the `if`, `elif`, or `else` to be bundled together by a tab (or 4 spaces) at the start of the line. Fierce discussions have been had between programmers about whether using a tab or 4 spaces is better (I wish I was kidding...), but both are fine in my book. As long as you pay very close attention to the spacing.

```python
## Code block 8: Pay attention the spacing

if 3 < 4:
    print('3 is indeed less than 4')
    print('That speaks for itself')
    print('Why do you even ask such a question?')
    print('Are you bored or something?')
else:
    print('No, 3 is more than 4')
    print('Math has been revolutionized')
    print('Have you ever heard of the actor Terrence Howard?')
    print('He has invented a type of math where this might be true.')
    print('He was right all along.')
```

As you see here, there is kind of a "text island" below the if-statement and else-statement. There are multiple print statements that are all one tab up from the start of the line. This bundles all the print lines together below the if-statement, and else-statement respectively. Feel free to change the `3 < 4` with another comparison expression to also see the lines below `else` in action.
